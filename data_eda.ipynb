### New Columns

```python
# Time based columns
df['tiempo_espera'] = df['hora_llamado'] - df['hora_llegada']
df['tiempo_atencion'] = df['hora_salida'] - df['hora_llamado']
df['tiempo_total'] = df['hora_salida'] - df['hora_llegada']
```

```python
df.head()
```

### Data Exploration

```python
# Distribution of categorical variables
print("\Distribución por Segmento:\n", df['Segmento'].value_counts())
print("\Distribución por Tienda:\n", df['tienda'].value_counts())
print("\nStatus Count:\n", df['status'].value_counts())
print("\nState Count:\n", df['estado'].value_counts())
```

#### Graphics

```python
df.columns
```

```python
df['status'].value_counts().plot(kind='bar')
plt.title('Status Distribution')
plt.xlabel('Status')
plt.show()

df['Segmento'].value_counts().plot(kind='bar')
plt.title('Segmento Distribution')
plt.xlabel('Segmento')
plt.show()
```

### Data Sample

```python
# Definir tamaño de la muestra
sample_frac = 0.1

# Aplicar muestreo estratificado manualmente
df_sample = df.groupby(['estado', 'tienda'], group_keys=False).apply(lambda x: x.sample(frac=sample_frac, random_state=42))

# Resetear índices
df_sample = df_sample.reset_index(drop=True)
```

```python
df_sample.to_csv("/content/drive/MyDrive/Colab Notebooks/sample_data.csv", index=False)
```
